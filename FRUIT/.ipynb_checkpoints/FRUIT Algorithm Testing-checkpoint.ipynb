```python
def thermal(E, T_0):
    return ( (E / (np.power(T_0, 2)))*np.exp(-E/T_0) ) * (E < 1e-7)

def epithermal(E, E_d, b, beta_p):
    return ( (1 - np.exp( -np.power((E/E_d), 2) )) * np.power(E, (b-1)) * np.exp(-E/beta_p) ) * (E < 0.1) * (E > 1e-7)

def fast_fission(E, alpha, beta):
    return (np.power(E, alpha)) * np.exp(-E / beta)

def fast_evaporation(E, T_ev):
    return (E / (np.power(T_ev, 2))) * np.exp(-E / T_ev) 

def fast_gaussian(E, E_m, sigma):
    return np.exp( - np.power((E - E_m), 2) / (2 * np.power((sigma * E_m), 2) ) )

def fast_high_energy(E, T_ev):
    return (E / (np.power(T_ev, 2))) * np.exp(-E / T_ev)

def high_energy_func(E, T_hi):
    return (E / (np.power(T_hi, 2))) * np.exp(-E / T_hi)
```

```python
def fission(E, P_th, P_e, T_0, E_d, b, beta_p, alpha, beta):
    E = E * 1e-6
    while P_th + P_e > 1:
        P_th = P_th / 2.0
        P_e = P_e / 2.0
    P_f = 1 - P_th - P_e
    
    spectra = P_th * thermal(E, T_0) + P_e * epithermal(E, E_d, b, beta_p) + P_f * fast_fission(E, alpha, beta)
    return spectra / np.sum(spectra)

def evaporation(E, P_th, P_e, T_0, E_d, b, beta_p, T_ev):
    E = E * 1e-6
    while P_th + P_e > 1:
        P_th = P_th / 2.0
        P_e = P_e / 2.0
    P_f = 1 - P_th - P_e
    
    spectra = P_th * thermal(E, T_0) + P_e * epithermal(E, E_d, b, beta_p) + P_f * fast_evaporation(E, T_ev)
    return spectra / np.sum(spectra)

def gaussian(E, P_th, P_e, T_0, E_d, b, beta_p, E_m, sigma):
    E = E * 1e-6
    while P_th + P_e > 1:
        P_th = P_th / 2.0
        P_e = P_e / 2.0
    
    P_f = 1 - P_th - P_e
    
    spectra = P_th * thermal(E, T_0) + P_e * epithermal(E, E_d, b, beta_p) + P_f * fast_gaussian(E, E_m, sigma)
    return spectra / np.sum(spectra)

def high_energy(E, P_th, P_e, P_f, T_0, E_d, b, beta_p, T_ev, T_hi):
    E = E * 1e-6
    while P_th + P_e + P_f > 1:
        P_th = P_th / 2.0
        P_e = P_e / 2.0
        P_f = P_f / 2.0
    P_he = 1 - P_th - P_e - P_f
    
    spectra = P_th * thermal(E, T_0) + P_e * epithermal(E, E_d, b, beta_p) + P_f * fast_high_energy(E, T_ev) + P_he * high_energy_func(E, T_hi)
    return spectra / np.sum(spectra)
```

```python
def plot_loghist(ax, x, bins):
  hist, bins = np.histogram(x, bins=bins)
  logbins = np.logspace(np.log10(bins[0]),np.log10(bins[-1]),len(bins))
  ax.hist(x, bins=logbins)
  ax.set_xscale('log')

fittings = fruit_fittings.FRUIT_FITTINGS
```

```python
# Making Histogram of Fission Spectra Parameter Fitting

P_th, P_e, T_0, E_d, b, beta_p, alpha, beta = [], [], [], [], [], [], [], []
for fit in fittings:
    
    if fit["fit_type"] in "fission":
        while (fit["params"][0] + fit["params"][1]) > 1.0:
            fit["params"][0] = fit["params"][1] / 2.0
            fit["params"][1] = fit["params"][1] / 2.0
        P_th.append(fit["params"][0]) 
        P_e.append(fit["params"][1]) 
        T_0.append(fit["params"][2]) 
        E_d.append(fit["params"][3]) 
        b.append(fit["params"][4]) 
        beta_p.append(fit["params"][5]) 
        alpha.append(fit["params"][6]) 
        beta.append(fit["params"][7]) 
        
        
```

```python
fig, axs = plt.subplots(3,3, figsize=(14,14))
BINS = 15

axs[0][0].hist(P_th, bins = BINS)
axs[0][0].set_xlabel("Thermal proportion")
axs[0][0].set_ylabel("Frequency")

axs[0][1].hist(P_e, bins = BINS)
axs[0][1].set_xlabel("Epithermal proportion")
axs[0][1].set_ylabel("Frequency")

axs[0][2].hist(1.0 - np.array(P_th) - np.array(P_e), bins = BINS)
axs[0][2].set_xlabel("Fast proportion")
axs[0][2].set_ylabel("Frequency")

plot_loghist(axs[1][0], T_0, bins = BINS)
axs[1][0].set_xlabel("T_0 (eV)")
axs[1][0].set_ylabel("Frequency")

axs[1][1].hist(E_d, bins = BINS, range = (0, 1e-6))
axs[1][1].set_xlabel("E_d (eV)")
axs[1][1].set_ylabel("Frequency")

axs[1][2].hist(b, bins = BINS, range = (0, 3))
axs[1][2].set_xlabel("b")
axs[1][2].set_ylabel("Frequency")

plot_loghist(axs[2][0], beta_p, bins = BINS)
axs[2][0].set_xlabel("beta_p")
axs[2][0].set_ylabel("Frequency")

axs[2][1].hist(alpha, bins = BINS, range = (0, 4))
axs[2][1].set_xlabel("alpha")
axs[2][1].set_ylabel("Frequency")

plot_loghist(axs[2][2], beta, bins = BINS)
axs[2][2].set_xlabel("beta")
axs[2][2].set_ylabel("Frequency")

fig.suptitle("Fitting parameters for IAEA data which were best fit by a fission spectra")
plt.tight_layout()

```
